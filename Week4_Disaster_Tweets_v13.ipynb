**Data Cleaning**

```python
import re
import nltk
from nltk.corpus import stopwords

def text_cleaning(text) :
    text = re.sub(r"http\S+","", text)
    text = re.sub(r"[^A-Za-z\s]", "", text)
    text = text.lower()
    text = text.split()
    text = [word for word in text if word not in stopwords.words("english")]
    text = " ".join(text)
    return text
    
train_data["cleaned_text"] = train_data["text"].apply(text_cleaning)
test_data["cleaned_text"] = test_data["text"].apply(text_cleaning)
```

```python
print(train_data[["text", "cleaned_text"]].head())
```

**Model Architecture**

```python
tokenizer = Tokenizer(num_words = 5000, lower = True)
tokenizer.fit_on_texts(train_data["cleaned_text"])

X = tokenizer.texts_to_sequences(train_data["cleaned_text"])
X = pad_sequences(X, maxlen = 50)

y= train_data["target"].values
```

**LSTM Model and Optimization**

I am going to use a LSTM model for the classification.

I will process with Bayesian optimization with Kfold cross validation and early_stopping to help prevent over_fitting.

```python
class F1ScoreCallback(tf.keras.callbacks.Callback) :
    
    def __init__(self, validation_data):
        super(F1ScoreCallback, self).__init__()
        self.validation_data = validation_data
    
    def on_epoch_end(self, epoch, logs = None):
        val_data = self.validation_data
        y_val_pred = (self.model.predict(val_data[0]) > 0.5).astype("int32")
        y_val_true = val_data[1]
        
        f1 = f1_score(y_val_true, y_val_pred)
        
        print(f"- val_f1 : {f1:.4f}")
        
        if logs is not None:
            logs["val_f1"] = f1       

def build_model(hp) : 
    model = tf.keras.Sequential([
    tf.keras.layers.Embedding(input_dim = 10000, output_dim = 128),
    tf.keras.layers.LSTM(hp.Int("units", 32, 128, 32)),
    tf.keras.layers.Dense(1, activation = "sigmoid")
    ])

    model.compile(
        optimizer = tf.keras.optimizers.Adam(hp.Float("learning_rate", min_value = 1e-6, max_value = 1e-2, sampling = "log")),
        loss = "binary_crossentropy", 
        metrics = ["accuracy"]
    )
    
    return model


kfold = KFold(n_splits = 5, shuffle = True, random_state = 42)
best_val_f1 = -np.inf

for fold_nb, (train_index, val_index) in enumerate(kfold.split(X, y), 1):
    shutil.rmtree(f"tuner_dir_fold_{fold_nb}", ignore_errors = True)
    print(f"Training on fold {fold_nb}...")
    X_train_fold, X_val_fold = X[train_index], X[val_index]
    y_train_fold, y_val_fold = y[train_index], y[val_index]
    
    tuner = kt.BayesianOptimization(
        build_model,
        objective = kt.Objective("val_f1", direction = "max"),
        max_trials = 8,
        directory = f"tuner_dir_fold_{fold_nb}",
        project_name = "Disaster_Tweets_tuner_v13"
    )
    
    f1_callback = F1ScoreCallback(validation_data=(X_val_fold, y_val_fold))
    
    early_stopping = EarlyStopping(
        monitor = "val_f1", 
        patience = 2, 
        verbose = 1, 
        mode = "max"
    )
    
    tuner.search(
        X_train_fold, y_train_fold,
        epochs = 8,
        validation_data = (X_val_fold, y_val_fold),
        callbacks = [early_stopping, f1_callback],
        batch_size = 32,
        verbose = 2
    )
    
    best_model = tuner.get_best_models(num_models = 1)[0]

    y_val_pred = (best_model.predict(X_val_fold) > 0.5).astype("int32")
    y_val_true = y_val_fold
    
    val_f1 = f1_score(y_val_true, y_val_pred)

    print(f"Validation F1 Score for fold {fold_nb}: {val_f1: .4f}")
    
    if val_f1 > best_val_f1 :
        best_val_f1 = val_f1
        best_model.save("Disaster_Tweets_best_model_v13.keras")
        
print(f"Best validation F1 Score: {best_val_f1 : .4f}")
```

```python
from keras.models import load_model
loaded_model = load_model("Disaster_Tweets_best_model_v13.keras")
```

```python
X_test = tokenizer.texts_to_sequences(test_data["cleaned_text"])
X_test = pad_sequences(X_test, maxlen = 50)

predictions = loaded_model.predict(X_test)
predicted_classes = (predictions > 0.5).astype("int32")
```
